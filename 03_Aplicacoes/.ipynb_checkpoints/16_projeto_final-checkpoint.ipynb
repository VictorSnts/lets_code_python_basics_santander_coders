```python
import requests as rq
import datetime as dt
import csv
from PIL import Image
from IPython.display import display
```

```python
# Importando os dados brutos da API
url = "https://api.covid19api.com/dayone/country/brazil"
response = rq.get(url)
response.status_code
```

```python
# Transformando os dados brutos em JSON
dados_brutos = response.json()
dados_brutos
```

```python
# Filtrando os dados 
dados = [["Confirmed", "Deaths", "Recovered", "Active", "Date"]]
for info in dados_brutos:
    dados.append([info["Confirmed"], info["Deaths"], info["Recovered"], info["Active"], info["Date"]])
dados
```

```python
# Criando constantes para facilitar a manipulacao
CONFIRMED = 0
DEATHS = 1
RECOVERED = 2
ACTIVE = 3
DATE = 4
```

```python
# Alterando o formato da data de timestamp para formato AAAA-MM-DD
for registro in range(1, len(dados)):
    dados[registro][DATE] = dados[registro][DATE][:10]
dados
```

```python
# Salvando o arquivo em CSV
with open("brasil-covid19.csv", 'w') as arquivo:
    writer = csv.writer(arquivo)
    writer.writerow(dados)
```

```python
# Alterando a string da data para o tipo data
for registro in range(1, len(dados)):
    dados[registro][DATE] = dt.datetime.strptime(dados[registro][DATE], "%Y-%m-%d")
dados
```

```python
def get_datasets(y, labels):
    if type(y[0]) == list:
        datasets = []
        for i in range(len(y)):
            datasets.append({
                "label": labels[i],
                "data": y[i]
            })
        return datasets
    else:
        return [{
                "label": labels[0],
                "data": y
            }]
```

```python
def set_title(title=""):
    if title != "":
        display = "true"
    else:
        display = "false"
    return {
        "title": title,
        "display": display
    }
```

```python
def create_chart(x, y, labels, kind="bar", title=""):
    datasets = get_datasets(y, labels)
    options = set_title(title)
    
    chart = {
        "type": kind,
        "data": {
            "labels": x,
            "datasets": datasets
      },
        'options': options
    }
    return chart
```

```python
def get_api_chart(chart):
    url_base = 'https://quickchart.io/chart'
    requisicao = rq.get(f'{url_base}?c={str(chart)}')
    return requisicao.content
```

```python
def save_image(path, content):
    with open(path, 'wb') as image:
        image.write(content)
```

```python
def display_image(path):
    img_pil = Image.open(path)
    display(img_pil)
```

```python
y_data_1 = []
for obs in dados[1::10]:
    y_data_1.append(obs[CONFIRMED])

y_data_2 = []
for obs in dados[1::10]:
    y_data_2.append(obs[RECOVERED])

labels = ['Confirmados', 'Recuperados']

print(y_data_2)

x = []

for obs in dados[1::10]:
    x.append(obs[DATE].strftime('%d/%m/%a'))

chart = create_chart(x, [y_data_1, y_data_2], labels, title='Gráfico confirmado x recuperados')
chart_content = get_api_chart(chart)
save_image('meu-primeiro-grafico.png', chart_content)
display_image('meu-primeiro-grafico.png')
```
